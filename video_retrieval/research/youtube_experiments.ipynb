```python
# Replace 'YOUR_API_KEY' with the API key you obtained from the Google Cloud Console
api_key = ''
youtube = build('youtube', 'v3', developerKey=api_key)

def search_videos(query, max_results=5):
    request = youtube.search().list(
        part='snippet',
        q=query,
        type='video',
        maxResults=max_results
    )

    response = request.execute()
    return response['items']

if __name__ == "__main__":
    search_query = input("Enter your search query: ")
    results = search_videos(search_query)

    for video in results:
        video_id = video['id']['videoId']
        video_title = video['snippet']['title']
        print(f"{video_title} - https://www.youtube.com/watch?v={video_id}")
```

```python
import os
from googleapiclient.discovery import build

# Replace 'YOUR_API_KEY' with your actual YouTube Data API key
API_KEY = ''

def get_video_captions(video_id):
    youtube = build('youtube', 'v3', developerKey=API_KEY)

    # Get captions list for the video
    captions_response = youtube.captions().list(
        part='snippet',
        videoId=video_id
    ).execute()

    # Check if captions exist for the video
    if 'items' in captions_response:
        captions_items = captions_response['items']
        if captions_items:
            # Get the first caption track (usually the automatic captions)
            caption_track_id = captions_items[0]['id']
            
            # Retrieve the actual caption content
            caption_content = youtube.captions().download(
                id=caption_track_id,
                tfmt='srt'  # SRT format for captions
            ).execute()

            return caption_content
    return None

if __name__ == "__main__":
    video_id = 'ED9EU3FfzyU'
    captions = get_video_captions(video_id)

    if captions:
        # Save the captions to a file
        output_file = f'{video_id}_captions.srt'
        with open(output_file, 'wb') as f:
            f.write(captions)
        print(f"Captions saved to '{output_file}'.")
    else:
        print("Captions not available for this video.")
```
